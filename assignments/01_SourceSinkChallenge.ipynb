```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.dpi'] = 100
rcParams['font.size'] = 16
rcParams['font.family'] = 'StixGeneral'
```

```python
N = 100                               # number of points in each direction
x_start, x_end = -2.0, 2.0            # boundaries in the x-direction
y_start, y_end = -1.0, 1.0            # boundaries in the y-direction
x = np.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = np.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = np.meshgrid(x, y)              # generates a mesh grid
plt_den = 30                          # contour density
```

### Source

```python
strength_source = 5.0                      # source strength
x_source, y_source = -1.0, 0.0             # location of the source
```

```python
# calculating the potential of a source
phi_source = strength_source/(2*pi) * np.log1p(np.sqrt((X-x_source)*(X-x_source) + (Y-y_source)*(Y-y_source)))
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
fig1 = plt.contour(X, Y, phi_source, plt_den)
plt.clabel(fig1, inline=1, fontsize=10)
plt.show();
```

### Sink

```python
strength_sink = -5.0                     # strength of the sink
x_sink, y_sink = 1.0, 0.0                # location of the sink
```

```python
# calculating the potential of a sink
phi_sink = strength_sink/(2*pi) * np.log1p(np.sqrt((X-x_sink)*(X-x_sink) + (Y-y_sink)*(Y-y_sink)))
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
fig2 = plt.contour(X, Y, phi_sink, plt_den)
plt.clabel(fig2, inline=1, fontsize=10)
plt.show();
```

### Source-sink pair

```python
phi_pair = strength_source/(2*pi) * np.log1p(np.sqrt((X-x_source)*(X-x_source) + (Y-y_source)*(Y-y_source))) \
+ strength_sink/(2*pi) * np.log1p(np.sqrt((X-x_sink)*(X-x_sink) + (Y-y_sink)*(Y-y_sink)))
```

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
fig3 = plt.contour(X, Y, phi_pair, plt_den)
plt.clabel(fig3, inline=1, fontsize=10)
plt.show();
```
