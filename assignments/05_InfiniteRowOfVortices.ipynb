##### Finite Row of Vortices

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 51                                # number of points in each direction
x_start, x_end = -2.0, 2.0            # boundaries in the x-direction
y_start, y_end = -0.5, 0.5            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
n_vortex = 201    # number of vortex, keep it odd
a = 0.4
x_vortex = numpy.linspace(-(n_vortex-1)*a/2.,(n_vortex-1)*a/2.,n_vortex)
y_vortex = numpy.zeros(n_vortex)
gamma = 10.0                      # strength of the vortex
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    u = + strength/(2*numpy.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*numpy.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u, v
```

```python
u_total = numpy.zeros((N,N))
v_total = numpy.zeros((N,N))
for i in range(n_vortex):
    u_temp, v_temp = get_velocity_vortex(gamma, x_vortex[i], y_vortex[i], X, Y)
    u_total += u_temp
    v_total += v_temp
```

```python
size = 14
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=20)
pyplot.ylabel('y', fontsize=20)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_total, v_total, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vortex, y_vortex, color='r', s=70, marker='o');
```

##### Infinite row of vortices

```python
u_inf = gamma/(2*a)*numpy.sinh(2*numpy.pi*Y/a)/(numpy.cosh(2*numpy.pi*Y/a)-numpy.cos(2*numpy.pi*X/a))
v_inf = -gamma/(2*a)*numpy.sin(2*numpy.pi*X/a)/(numpy.cosh(2*numpy.pi*Y/a)-numpy.cos(2*numpy.pi*X/a))
```

```python
size = 14
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=20)
pyplot.ylabel('y', fontsize=20)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_inf, v_inf, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vortex, y_vortex, color='r', s=70, marker='o');
```
