###### Circle Geometry

```python
r = 1.2    # circle radius
n = 50     # number of points on the circle
xi_c, eta_c = -0.15, 0.     # center of the circle
xi_0 = numpy.linspace(xi_c-r,xi_c+r,n+1)   # x coordinates: upper circle
xi = numpy.concatenate((xi_0,xi_0))        # x coordinates: whole circle
eta_0 = numpy.sqrt(r**2-(xi_0-xi_c)**2) + eta_c    # y coordinates: upper circle
eta_1 = -numpy.sqrt(r**2-(xi_0-xi_c)**2) + eta_c   # y coordinates: lower circle
eta = numpy.concatenate((eta_0, eta_1))    # y coordinates: whole circle
```

```python
zeta = xi + 1j * eta     # circle geom complex
```

```python
pyplot.plot(xi,eta,'x')
pyplot.axis('equal');
```

###### Joukowski Transform >> Airfoil Geom

```python
z = zeta + 1/zeta   # Joukowski Trans
```

```python
x_airfoil = z.real   # x coord: Airfoil
y_airfoil = z.imag   # y coord: Airfoil
```

```python
pyplot.plot(x_airfoil,y_airfoil,'x')
pyplot.axis('equal');
```

###### Meshgrid Gen: Circle

```python
N = 50                                # Number of points in each direction
x_start, x_end = -5.0, 5.0            # x-direction boundaries
y_start, y_end = -3.0, 3.0            # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # computes a 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # computes a 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid
```

```python
# plots the grid of points
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.plot(xi,eta,'x')
pyplot.scatter(X, Y, s=10, color='#CD2305', marker='o', linewidth=0);
```

###### Meshgrid Trans >> Airfoil Plane Meshgrid

```python
grid_cir = X + 1.0j*Y
```

```python
grid_airfoil = grid_cir + 1.0/grid_cir
grid_air_x = grid_airfoil.real
grid_air_y = grid_airfoil.imag
```

```python
# plots the grid of points
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.plot(x_airfoil,y_airfoil,'x',linewidth=5)
pyplot.scatter(grid_air_x, grid_air_y, s=10, color='#CD2305', marker='o', linewidth=0);
```

###### Potential field configuration

```python
u_inf = 1.0     # freestream speed
kappa = r**2*2*numpy.pi*u_inf     # strength of the doublet
xi_doublet, eta_doublet = xi_c, eta_c    # location of the doublet
gamma = 0.0     # strength of the vortex
xi_vortex, eta_vortex = xi_c, eta_c    # location of the vortex
```

###### Calculate the potential field on *Circle*

```python
# computes the velocity field on the mesh grid
u_doublet, v_doublet = get_velocity_doublet(kappa, xi_doublet, eta_doublet, X, Y)
# computes the stream-function on the mesh grid
psi_doublet = get_stream_function_doublet(kappa, xi_doublet, eta_doublet, X, Y)

# computes the velocity field on the mesh grid
u_vortex, v_vortex = get_velocity_vortex(gamma, xi_vortex, eta_vortex, X, Y)
# computes the stream-function on the mesh grid
psi_vortex = get_stream_function_vortex(gamma, xi_vortex, eta_vortex, X, Y)

# freestream velocity components
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)
# stream-function of the freestream flow
psi_freestream = u_inf * Y

# superposition of the doublet and the vortex on the freestream flow
u = u_freestream + u_doublet + u_vortex
v = v_freestream + v_doublet + v_vortex
psi = psi_freestream + psi_doublet + psi_vortex
```

```python
# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1.5, arrowstyle='->')
circle = pyplot.Circle((xi_c, eta_c), radius=r, color='#CD2305', alpha=0.5)
pyplot.gca().add_patch(circle);
```

###### Transform Streamline into new coords

```python
vel = u + 1.0j*v
air_vel = vel + 1.0/vel
u_airfoil = air_vel.real
v_airfoil = air_vel.imag
```

```python
# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
max_psi,min_psi = numpy.max(psi),numpy.min(psi)
levels = numpy.linspace(min_psi,max_psi, 100)
pyplot.contour(grid_air_x, grid_air_y, psi, levels=levels)
pyplot.plot(x_airfoil,y_airfoil,'o',markersize=5,markerfacecolor='red', alpha = 0.5);
```
